### CONVERTING skater_summary.csv to JSON

```python
import pandas as pd
from pathlib import Path

# === FILE PATHS ===
INPUT_PATH = Path(r"C:\Users\cadla\Desktop\Documents\2026-TopShelfAnalytics\topshelfanalytics-skater-table\skater_summary.csv")
OUTPUT_PATH = Path(r"C:\Users\cadla\Desktop\Documents\2026-TopShelfAnalytics\topshelfanalytics-skater-table\skater_summary.json")

# === LOAD CSV ===
df = pd.read_csv(INPUT_PATH)

# === OPTIONAL CLEANING ===

# Convert datetime columns to clean YYYY-MM-DD (adjust column names if needed)
for col in df.columns:
    if "date" in col.lower():
        try:
            df[col] = pd.to_datetime(df[col], errors="coerce").dt.strftime("%Y-%m-%d")
        except Exception:
            pass

# Replace NaN with None (better for JSON)
df = df.where(pd.notnull(df), None)

# === EXPORT JSON ===
df.to_json(
    OUTPUT_PATH,
    orient="records",   # list of row objects (best for frontend)
    indent=2            # pretty format (optional)
)

print(f"✅ JSON file created: {OUTPUT_PATH.resolve()}")
```
